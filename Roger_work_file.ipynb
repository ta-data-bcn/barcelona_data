# Name trends

# Baby names through 1996-2017

```python
baby_df = pd.read_csv('names/most_frequent_baby_names.csv')
```

```python
baby_df.head()
```

```python
# Checking top 10 names for women in every year

for year in set(baby_df['Year']):
    print(baby_df[baby_df['Year'] == year].head(10))
```

```python
# Defining new data frame for men
baby_df_men = baby_df[baby_df['Gender'] == 'Male']
```

```python
# Checking top 10 names for men in every year

for year in set(baby_df_men['Year']):
    print(baby_df_men[baby_df_men['Year'] == year].head(10))
```

```python
# Creating series with top names for all years (MEN)

common_baby_names_men = baby_df[baby_df['Gender'] =='Male'].groupby('Name').sum()['Frequency']
common_baby_names_men.sort_values(ascending=False).head(10)
```

```python
# Creating series with top names for all years (WOMEN)
common_baby_names_women = baby_df[baby_df['Gender'] =='Female'].groupby('Name').sum()['Frequency']
common_baby_names_women.sort_values(ascending=False).head(10)
```

```python
# Creating dictionary of dataframes for top 10 baby names for every year in order to see what names appear the most (MEN)
dict_men = {}
for year in set(baby_df_men['Year']):
    dict_men[year] = baby_df_men[baby_df_men['Year'] == year].head(10)
```

```python
dict_men
```

```python
# Concatenating all dataframes in one
result_men = pd.concat([dict_men[i] for i in dict_men.keys()])
result_men
```

```python
# Times that a name has appeared in the top 10 (MEN)
result_men['Name'].value_counts()
```

```python
# Creating dictionary of dataframes for top 10 baby names for every year in order to see what names appear the most (WOMEN)
dict_women = {}
for year in set(baby_df['Year']):
    dict_women[year] = baby_df[baby_df['Year'] == year].head(10)
```

```python
dict_women
```

```python
# Concatenating all dataframes in one
result_women = pd.concat([dict_women[i] for i in dict_women.keys()])
result_women
```

```python
# Times that a name has appeared in the top 10 (WOMEN)
result_women['Name'].value_counts()
```

# Immigration

# By Nationality

```python
immigration_nat = pd.read_csv('immigrants_by_nationality/immigrants by nationality/immigrants_by_nationality.csv')
```

```python
immigration_nat.head()
```

```python
# Deleting Neighborhood Name and Code
del immigration_nat['Neighborhood Name']
del immigration_nat['Neighborhood Code']
```

```python
immigration_nat.head()
```

```python
# Top immigration over last 3 years
immigration_nat[['Nationality','Number']].groupby('Nationality').sum().sort_values(by='Number', ascending=False).head(15)
```

```python
immigration_nat[immigration_nat['Year'] == 2017][['Nationality','Number']].groupby('Nationality').sum().sort_values(by='Number', ascending=False).head(15)
```

```python
# Total immigrants in 2017

print('2017: ', immigration_nat[immigration_nat['Year'] == 2017][['Nationality','Number']].groupby('Nationality').sum().sort_values(by='Number', ascending=False).sum()[0])
print('2016: ', immigration_nat[immigration_nat['Year'] == 2016][['Nationality','Number']].groupby('Nationality').sum().sort_values(by='Number', ascending=False).sum()[0])
print('2015: ', immigration_nat[immigration_nat['Year'] == 2015][['Nationality','Number']].groupby('Nationality').sum().sort_values(by='Number', ascending=False).sum()[0])
```

```python
# Top immigration over last 3 years by district

immigration_nat[['District Name','Nationality','Number']].groupby(['District Name','Nationality']).sum().sort_values(by=['District Name','Number'], ascending=[True,False]).head()
```

```python
for district in set(immigration_nat['District Name']):
    if district == 'No consta':
        continue
    else:
        print(f'\n{district}')
        print(immigration_nat[immigration_nat['District Name'] == district][['District Name','Nationality','Number']].groupby('Nationality').sum().sort_values('Number', ascending=False).head(10))
```

```python
for year in set(immigration_nat['Year']):
    print(f'\n{year}')
    print(immigration_nat[immigration_nat['Year'] == year][['District Name','Nationality','Number']].groupby('Nationality').sum().sort_values('Number', ascending=False).head(15))
```

```python
nat_dict = {}
for year in set(immigration_nat['Year']):
    temp_df = immigration_nat[immigration_nat['Year'] == year][['District Name','Nationality','Number']].groupby('Nationality').sum().sort_values('Number', ascending=False).head(15)
    temp_df['Year'] = year
    nat_dict[year] = temp_df
```

```python
nat_dict
```

```python
nat_result = pd.concat([nat_dict[i] for i in nat_dict.keys()])
```

```python
nat_2_df = nat_result.sort_values('Nationality')
nat_2_df.reset_index(inplace=True)
```

```python
nat_2_df.sort_values(['Nationality','Year'], inplace=True)
```

```python
nat_2_df.to_csv('nat_1.csv')
```
