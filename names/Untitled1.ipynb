```python
import pandas as pd
```

```python
df = pd.read_csv('25_noms_padro_any_sexe_1996_2017.csv')
```

```python
df.head()
```

```python
for year in set(df['Any']):
    print(df[df['Any'] == year].head(10))
```

```python
df_male = df[df['Sexe'] == 'Home']
```

```python
for year in set(df_male['Any']):
    print(df_male[df_male['Any'] == year].head(10))
```

```python
common_names = df.groupby('Nom').sum()['Nombre']
```

```python
common_names.sort_values(ascending=False).head()
```

```python
common_names_women = df[df['Sexe'] =='Dona'].groupby('Nom').sum()['Nombre']
```

```python
common_names_women.sort_values(ascending=False).head(10)
```

```python
common_names_men = df[df['Sexe'] =='Home'].groupby('Nom').sum()['Nombre']
```

```python
common_names_men.sort_values(ascending=False).head(10)
```
