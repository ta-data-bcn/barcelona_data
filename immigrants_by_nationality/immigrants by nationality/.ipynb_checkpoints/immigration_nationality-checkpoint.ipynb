```python
import pandas as pd
```

```python
df = pd.read_csv('2017_immigrants_nacionalitats.csv')
```

```python
df.head()
```

```python
df.groupby('Nacionalitat').sum().sort_values(by='Nombre', ascending=False).head(10)
```

```python
df.groupby(['Nom_Districte','Nacionalitat']).sum().sort_values(['Nom_Districte','Nombre'],ascending=[True,False])
```

```python
df.head()
```

```python
for districte in set(df['Nom_Districte']):
    print('\n',districte)
    print(df[df['Nom_Districte']==districte][['Nacionalitat','Nombre']].groupby('Nacionalitat').sum().sort_values(by='Nombre',ascending=False).head())
```
