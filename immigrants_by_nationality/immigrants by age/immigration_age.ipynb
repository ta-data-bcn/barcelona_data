```python
import pandas as pd
```

```python
df = pd.read_csv('2017_immigrants_edats-quinquennals.csv')
```

```python
df.head()
```

```python
df.groupby('Edat_quinquennal').sum().sort_values(by='Nombre',ascending=False)
```

```python
df.groupby(['Nom_Districte', 'Edat_quinquennal']).sum().sort_values(['Nom_Districte', 'Nombre'], ascending=[True,False])
```

```python
for districte in set(df['Nom_Districte']):
    print('\n',districte)
    print(df[df['Nom_Districte']==districte][['Edat_quinquennal','Nombre']].groupby('Edat_quinquennal').sum().sort_values(by='Nombre',ascending=False).head())
```
