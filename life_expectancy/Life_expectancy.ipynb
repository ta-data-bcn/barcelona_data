```python
import pandas as pd
import numpy as np
```

```python
l_exp = pd.read_csv("Life_expectancy_of_the-city_of_Barcelona.csv")

l_exp_small = l_exp[["Dte.", "Barris", "2007-2011", "2010-2014"]]
l_exp_small.columns = ["Districte", "Barris", "2007-2011", "2010-2014"]
l_exp_small.head()
```

```python
l_exp_small.dtypes
```

```python
l_exp_w = pd.read_csv("Life_expectancy_in_women_of_the_city_of_Barcelona.csv")
l_exp__wsmall = l_exp_w[["Dte.", "Barris", "2007-2011", "2010-2014"]]
l_exp__wsmall.columns = ["Districte", "Barris", "2007-2011W", "2010-2014W"]
l_exp__wsmall.head()
```

```python
l_exp_m = pd.read_csv("Life_expectancy_in_men_of_the_city_of_Barcelona.csv")
l_exp__msmall = l_exp_m[["Dte.", "Barris", "2007-2011", "2010-2014"]]
l_exp__msmall.columns = ["Districte", "Barris", "2007-2011M", "2010-2014M"]
l_exp__msmall.head()
```

```python
explore_exp = pd.concat([l_exp_small, l_exp__wsmall, l_exp__msmall], axis=1, sort=False)
```

```python
explore_exp = l_exp_small.append(l_exp__wsmall, sort=False)
```

```python
exp_merge = pd.merge(l_exp_small, l_exp__wsmall, on=['Barris', 'Barris'])
```

```python
exp_merge1 = pd.merge(exp_merge, l_exp__msmall, on=['Barris', 'Barris'])
```

```python
exp_merge1_small = exp_merge1[["Districte_x", "Barris", "2007-2011", "2010-2014", "2007-2011W", "2010-2014W", "2007-2011M", "2010-2014M"]]
```

```python
exp_merge1_small.sort_values('2010-2014', ascending = [False]).head()
```

```python
exp_merge1_small_dist = exp_merge1_small.groupby('Districte_x').mean().sort_values('2010-2014', ascending = [False])
```
