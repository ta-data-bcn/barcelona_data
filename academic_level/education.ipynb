```python
import pandas as pd
```

```python
df_18 = pd.read_csv('2018_padro_nivell_academic.csv')
```

```python
df_18.head()
```

```python
df_18.groupby('Nivell acadèmic').sum()
```

```python
# Level of education by neighborhood (total)
df_18.groupby(['Nivell acadèmic', 'Nom_Districte']).sum()
```

```python
# Level of education by neighborhood (men)
df_18[df_18['Sexe'] == 'Home'].groupby(['Nivell acadèmic', 'Nom_Districte']).sum()
```

```python
# Level of education by neighborhood (women)
df_18[df_18['Sexe'] == 'Dona'].groupby(['Nivell acadèmic', 'Nom_Districte']).sum().sort_values(by=['Nivell acadèmic','Nombre'],ascending=[True,False])
```

```python
# Level of education by neighborhood (women)
df_18[df_18['Sexe'] == 'Dona'].groupby(['Nom_Districte','Nivell acadèmic']).sum().sort_values(by=['Nom_Districte','Nombre'],ascending=[True,False])
```
