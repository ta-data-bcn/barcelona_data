```python
import pandas as pd
import numpy as np
```

```python
household = pd.read_csv("2018_padro_ocupacio_mitjana.csv")
household.head(5)
```

```python
household.dtypes
```

```python
household["Codi_Districte"].astype(object, copy=False)
household["Codi_Barri"].astype(object)
household["Any"].astype(object)

household["Codi_Districte"] = household["Codi_Districte"].astype('object')
household["Codi_Barri"] = household["Codi_Barri"].astype('object')
household["Any"] = household["Any"].astype('object')
```

```python
household.describe()
```

```python
household.groupby('Nom_Districte').mean().sort_values('Ocupació mitjana (persones per domicili)', ascending = [False])
```

```python
household.groupby('Nom_Barri').mean().sort_values('Ocupació mitjana (persones per domicili)', ascending = [False]).head(5)
```

```python
household.groupby('Nom_Barri').mean().sort_values('Ocupació mitjana (persones per domicili)', ascending = [False]).tail(5)
```
